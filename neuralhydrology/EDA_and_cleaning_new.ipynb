## import

```python
import pandas as pd
```

## cleaning functions

```python
def clean_daily(df):
    df.columns = df.iloc[0]
    df = df[3:]
    df.columns = df.columns.str.strip()
    df = df.drop(columns=['Ordinate'])
    df = df.rename(columns={'Date': 'Day', 'Time': 'Time'})
    df['Time'] = df['Time'].replace('24:00:00', '00:00:00')
    df['date'] = pd.to_datetime(df['Day'], format='%d-%b-%y') + pd.to_timedelta(df['Time'])
    df.dropna(subset=['date'], inplace=True)
    df.set_index('date', inplace=True)
    return df
```

```python
def clean_hourly(df):
    df.columns = df.iloc[0]
    df = df[3:]
    df.columns = df.columns.str.strip()
    df = df.drop(columns=['Ordinate'])
    df = df.rename(columns={'Date': 'Day', 'Time': 'Time'})
    #df['Time'] = df['Time'] + ":00"
    df['Time'] = df['Time'].replace('24:00:00', '00:00:00')
    df['date'] = pd.to_datetime(df['Day'], format='%d-%b-%y') + pd.to_timedelta(df['Time'])
    df.dropna(subset=['date'], inplace=True)
    df.set_index('date', inplace=True)
    return df
```

## check if clean daily works:

```python
daily_df = pd.read_csv("F:/Dino/UCB-USACE-LSTMs/data/updated_russian_river_folder/daily.csv", low_memory=False)
daily_df
```

```python
cleaned_daily = clean_daily(daily_df)
cleaned_daily
```

## check if clean hourly works:
