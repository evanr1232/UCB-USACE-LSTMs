```python
import pandas as pd
```

check if clean daily works:

```python
daily_df = pd.read_csv("/Users/mia/research/updated_russian_river_data_folder/daily.csv")
daily_df.head()
```

```python
def clean_daily(df):
    df.columns = df.iloc[0]
    df = df[3:]
    df.columns = df.columns.str.strip()
    df = df.drop(columns=['Ordinate'])
    df = df.rename(columns={'Date': 'Day', 'Time': 'Time'})
    df['Time'] = df['Time'].replace('24:00:00', '00:00:00')
    df['date'] = pd.to_datetime(df['Day'], format='%d-%b-%y') + pd.to_timedelta(df['Time'])
    df.set_index('date', inplace=True)
    return df
```

```python
cleaned_daily = clean_daily(daily_df)
cleaned_daily.head()
```

check if clean hourly works:

```python
hourly_df = pd.read_csv("/Users/mia/research/updated_russian_river_data_folder/hourly.csv")
hourly_df.head()
```

```python
def clean_hourly(df):
    df.columns = df.iloc[0]
    df = df[3:]
    df.columns = df.columns.str.strip()
    df = df.drop(columns=['Ordinate'])
    df = df.rename(columns={'Date': 'Day', 'Time': 'Time'})
    df['Time'] = df['Time'] + ":00"
    df['Time'] = df['Time'].replace('24:00:00', '00:00:00')
    df['date'] = pd.to_datetime(df['Day'], format='%d-%b-%y') + pd.to_timedelta(df['Time'])
    df.set_index('date', inplace=True)
    return df
```

```python
hourly_clean = clean_hourly(hourly_df)
hourly_clean.head()
```
