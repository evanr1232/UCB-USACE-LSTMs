```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
daily_df = pd.read_csv("/Users/mia/research/russian_river_data_folder/daily.csv")
print(daily_df.shape)
print(daily_df.columns)
daily_df.head()

#print(daily_df['date'].min(), daily_df['date'].max())
```

```python
def cleanDaily(df):
    df.columns = df.iloc[0]
    df = df[3:]
    df.columns = df.columns.str.strip()
    df = df.drop(columns=['Ordinate'])
    df = df.rename(columns={'Date': 'Day', '   Time': 'Time'})
    df['Time'] = df['Time'].replace('24:00:00', '00:00:00')
    df['date'] = pd.to_datetime(df['Day'], format='%d-%b-%y') + pd.to_timedelta(df['Time'])
    df.set_index('date', inplace=True)

    print(df.shape)
    print(df.index.min(), df.index.max())
    return df

daily_df = cleanDaily(daily_df)
print(daily_df.columns)
daily_df.head()
```

```python
# find all columns for guerneville:
keywords = [
    "BIG SULPHUR CR", "DRY CREEK 10", "GREEN VALLEY", "LAGUNA",
    "RUSSIAN 20", "RUSSIAN 30", "RUSSIAN 40", "RUSSIAN 50",
    "RUSSIAN 60", "RUSSIAN 70", "SANTA ROSA CR 10",
    "SANTA ROSA CR 20", "WF RUSSIAN"
]

# Function to filter columns
def filter_columns_by_keywords(df, keywords):
    filtered_columns = [col for col in df.columns if any(keyword in col for keyword in keywords)]
    return filtered_columns

# Example usage
filtered_columns = filter_columns_by_keywords(daily_df, keywords)
print("Filtered columns for guerneville:")
print(filtered_columns)
```

```python
#using calpella 2 which has an added row and
calpella_df = pd.read_csv("/Users/mia/research/russian_river_data_folder/calpella.csv")
print(calpella_df.shape)
calpella_df.head()
```

```python
def calpellaClean(df):
    df = df.copy()
    df = df[2:]
    df.columns = df.columns.str.strip()
    df = df.drop(columns=['Ordinate'])
    df = df.rename(columns={'Date / Time': 'date'})
    df['date'] = pd.to_datetime(df['date'], format='%d-%b-%y')
    df.set_index('date', inplace=True)
        
    print(df.shape)
    print(df.index.min(), df.index.max())
    return df
```

```python
calpella_df = calpellaClean(calpella_df)
calpella_df.head()
```

FOR THE REST OF THE PHYSICS INFORMED DATA SETS, THERE IS AN EXTRA ROW, SO THE CLEAN FUNCTION IS SLIGHTLY DIFFERENT FROM CALPELLA

```python
def physicsClean(df):
    df = df.copy()
    df = df[3:] #start at thirs row instead of 2nd like calpella
    df.columns = df.columns.str.strip()
    df = df.drop(columns=['Ordinate'])
    df = df.rename(columns={'Date / Time': 'date'})
    df['date'] = pd.to_datetime(df['date'], format='%d-%b-%y')
    df.set_index('date', inplace=True)
        
    print(df.shape)
    print(df.index.min(), df.index.max())
    return df
```

```python
hopland_df = pd.read_csv("/Users/mia/research/russian_river_data_folder/hopland.csv")
print(hopland_df.columns)
hopland_df.head()
```

```python
hopland_df_cleaned = physicsClean(hopland_df)
print(hopland_df_cleaned.shape)
print(hopland_df_cleaned.columns)
hopland_df_cleaned.head()
```

```python
warm_springs_df = pd.read_csv("/Users/mia/research/russian_river_data_folder/Warm Spring Inflow.csv")
print(warm_springs_df.columns)
warm_springs_df.head()
```

```python
warm_springs_df_cleaned = physicsClean(warm_springs_df)
print(warm_springs_df_cleaned.columns)
warm_springs_df_cleaned.head()
```

```python
#warm springs merged with daily:
warm_springs_physics_informed_df = pd.merge(daily_df, warm_springs_df_cleaned, left_index=True, right_index=True, how='outer')
print(warm_springs_physics_informed_df.shape)
print(warm_springs_physics_informed_df.index.min(), warm_springs_physics_informed_df.index.max())
print(warm_springs_physics_informed_df.columns)
warm_springs_physics_informed_df
```

```python
def is_invalid(value):
    # Check if the value is not NaN, not float, and not int
    return not pd.isna(value) and not isinstance(value, (float, int))
```
